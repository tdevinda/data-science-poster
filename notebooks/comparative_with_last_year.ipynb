```python
import pandas as pd
```

```python
immr_19 = pd.read_csv('../data/processed/immr_processed_2019.csv')
immr_18 = pd.read_csv('../data/processed/immr_processed_2018.csv')
```

```python
immr_19['total_male'] = immr_19['total_live_male'] + immr_19['total_dead_male']
immr_19['total_female'] = immr_19['total_live_female'] + immr_19['total_dead_female']

immr_18['total_male'] = immr_18['total_live_male'] + immr_18['total_dead_male']
immr_18['total_female'] = immr_18['total_live_female'] + immr_18['total_dead_female']
```

```python
group_sums_19 = immr_19[['disease_group', 'group_index', 'total_live_male', 'total_dead_male', 'total_live_female', 'total_dead_female']].groupby('disease_group').sum()

group_sums_18 = immr_18[['disease_group', 'group_index', 'total_live_male', 'total_dead_male', 'total_live_female', 'total_dead_female']].groupby('disease_group').sum()
```

```python
newcols = []
for col in group_sums_19:
    newcols.append(f'{col}_19')

group_sums_19.columns = newcols

newcols = []
for col in group_sums_18:
    newcols.append(f'{col}_18')

group_sums_18.columns = newcols
```

```python
group_sums_19.sort_values(by='group_index').head()
```

```python
group_sums_18.sort_values(by='group_index').head()
```

```python
ratios = group_sums_19.sort_values(by='group_index').values / group_sums_18.sort_values(by='group_index').values
group_sums_ratios = pd.DataFrame(ratios, columns=group_sums_19.columns, index=group_sums_19.index)
group_sums_ratios.head()
```
